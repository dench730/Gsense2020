```python
# def takemean( path ):
#     fh=fits.open(path)     
#     return  fh[0].data[1020:1060,850:950].mean()

# def takevariance( path ):
#     fh=fits.open(path)
#     return  fh[0].data[1020:1060,850:950].var()/2

# def takestd( path ):
#     fh=fits.open(path)
#     return  fh[0].data[1020:1060,850:950].std()

def takemean( path ):
    fh=fits.open(path)     
    return  fh[0].data[940:1000,990:1010].mean()

def takevariance( path ):
    fh=fits.open(path)
    return  fh[0].data[940:1000,990:1010].var()/2

def takestd( path ):
    fh=fits.open(path)
    return  fh[0].data[940:1000,990:1010].std()
```

```python
##[900:1000,980:1000]
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
import matplotlib.patches as patches

# 读取FITS文件
fits_file = "130nm-0007-0d03s-subdark.fit"
# fits_file = "78nm-0006-0d04s-subdark.fit"# 与74 72 ...比有移动
 

with fits.open(fits_file) as hdul:
    data = hdul[0].data

# 定义显示范围
x_min, x_max = 950, 1170
y_min, y_max = 600, 1200

# 定义高亮区域
hx_min, hx_max = 990, 1010
hy_min, hy_max = 940, 1000
 
# 提取子区域
subregion = data[y_min:y_max, x_min:x_max]

# 创建图形
fig, ax = plt.subplots(figsize=(10, 8))

# 显示图像
im = ax.imshow(subregion, 
               cmap='viridis', 
               origin='lower', 
               aspect='auto',
               extent=[x_min, x_max, y_min, y_max])

# 添加高亮区域的虚线框
rect = patches.Rectangle((hx_min, hy_min), 
                         hx_max - hx_min, 
                         hy_max - hy_min,
                         linewidth=2, 
                         linestyle='--', 
                         edgecolor='red',
                         facecolor='none')
ax.add_patch(rect)

# 设置坐标轴和标题
ax.set_xlabel('X Pixel', fontsize=12)
ax.set_ylabel('Y Pixel', fontsize=12)
ax.set_title(f'FITS Image: X({x_min}:{x_max}), Y({y_min}:{y_max})', fontsize=14)

# 添加网格
ax.grid(True, alpha=0.3)
# 添加颜色条
cbar = plt.colorbar(im, ax=ax)
cbar.set_label('ADU', fontsize=12)

plt.tight_layout()
plt.show()
```
